# Pneumonia Typing

- Focus on clinical outcomes of pneumonia and reports 

```python
import pandas as pd
import numpy as np 
```

```python
pneumonia_type_path = './pneumonia_type.csv'
full_cohort_processed_radgraph_extracts = '../../NER/full_cohort/processed_NER_extracts.csv'
full_cohort_clinical = '../../../cohort_data/cohort_data.csv'
pneumonia_type_df = pd.read_csv(pneumonia_type_path)
full_cohort_processed_radgraph_extracts_df = pd.read_csv(full_cohort_processed_radgraph_extracts)
full_cohort_clinical_df = pd.read_csv(full_cohort_clinical)
```

```python
pneumonia_type_df
print(pneumonia_type_df.shape)
# Y column indicates severity
```

```python
# Find study IDs that are not in full_cohort_radgraph_df

missing_study_ids = pneumonia_type_df.loc[
    ~pneumonia_type_df["study_id"].isin(full_cohort_processed_radgraph_extracts_df["study_id"]), "study_id"
]

# Print the count of missing study IDs
print(f"Total missing study IDs: {len(missing_study_ids)}")
# Display the first few missing study IDs
print(missing_study_ids.head())



missing_study_ids = pneumonia_type_df.loc[
    ~pneumonia_type_df["study_id"].isin(full_cohort_clinical_df["study_id"]), "study_id"
]

# Print the count of missing study IDs
print(f"Total missing study IDs: {len(missing_study_ids)}")
# Display the first few missing study IDs
print(missing_study_ids.head())
```

```python
# Merge Dataframes 

# Merge on 'study_id'
merged_df = pneumonia_type_df.merge(full_cohort_processed_radgraph_extracts_df, on="study_id", how="inner")
merged_df

# should have 4442 rows 
merged_df.rename(columns={'Y':'Severe'}, inplace=True)

# Save the merged DataFrame (optional)
merged_df.to_csv('pneumonia_type_radgraph.csv')
```
